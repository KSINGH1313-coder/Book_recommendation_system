```python
import ast
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
books['Book-Title'] = books['Book-Title'].apply(lambda x: " ".join(x.split()))
books['Book-Author'] = books['Book-Author'].apply(lambda x: " ".join(x.split()))
books['Publisher'] = books['Publisher'].apply(lambda x: " ".join(x.split()))
```

```python
books['Book-Title']=books['Book-Title'].apply(lambda x : x.replace(' ',''))
books['Book-Author']=books['Book-Author'].apply(lambda x : x.replace(' ',''))
books['Publisher']=books['Publisher'].apply(lambda x : x.replace(' ',''))
```

```python
books['Tags']=books['Book-Author']+books['Publisher']
```

```python
books = books.sample(1000)  # Limit to 1000 rows for testing
```

```python
new_df=books[['ISBN','Book-Title','Tags']]
```

```python
new_df.head()
```

```python
new_df['Tags']=new_df['Tags'].apply(lambda x : x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['Tags']).toarray()
```

```python
print(type(vectors))  # Should be a NumPy array or sparse matrix
print(vectors.shape)  # Should be (n_samples, n_features)
```

```python
vectors[0]
```

```python
import nltk
from nltk import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['Tags']=new_df['Tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)

```

```python
similarity[1]
```

```python
def recommend(book_title):
    book_title = book_title.strip().lower()

    # Ensure book exists in the dataset
    if book_title not in books['Book-Title'].str.lower().values:
        print("Book not found in the dataset! Please enter a valid title.")
        return

    # Get book index
    book_index = books[books['Book-Title'].str.lower() == book_title].index[0]

    # Check index bounds
    if book_index >= similarity.shape[0]:
        print("Book not found in similarity matrix. Try another book.")
        return

    # Get similarity scores
    similar_books = list(enumerate(similarity[book_index]))

    # Sort and exclude the same book
    similar_books = sorted(similar_books, key=lambda x: x[1], reverse=True)[1:6]

    print(f"Books similar to '{book_title}':")
    for i in similar_books:
        # Ensure book title has proper spacing
        book_name = books.iloc[i[0]]['Book-Title']
        print(" ".join(book_name.split()))  # Fix spacing issue
```
